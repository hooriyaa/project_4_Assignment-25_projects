```python
import turtle
import os  # For sound effects

# Set up the screen
wn = turtle.Screen()
wn.title("Pong Game ï¿½")  # Pong sticker
wn.bgcolor("black")
wn.setup(width=800, height=600)
wn.tracer(0)  # Stops the window from updating, we'll update manually

# Score
score_a = 0
score_b = 0

# Paddle A
paddle_a = turtle.Turtle()
paddle_a.speed(0)  # Maximum speed of animation
paddle_a.shape("square")
paddle_a.color("white")
paddle_a.shapesize(stretch_wid=5, stretch_len=1)  # Makes the paddle taller
paddle_a.penup()
paddle_a.goto(-350, 0)

# Paddle B
paddle_b = turtle.Turtle()
paddle_b.speed(0)
paddle_b.shape("square")
paddle_b.color("white")
paddle_b.shapesize(stretch_wid=5, stretch_len=1)
paddle_b.penup()
paddle_b.goto(350, 0)

# Ball
ball = turtle.Turtle()
ball.speed(0)
ball.shape("square")
ball.color("white")
ball.penup()
ball.goto(0, 0)
ball.dx = 0.2  # Ball's x movement speed
ball.dy = 0.2  # Ball's y movement speed

# Score display
pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0, 260)
pen.write("Player A: 0  Player B: 0", align="center", font=("Courier", 24, "normal"))

# Celebration sticker (displayed when someone scores)
celebration = turtle.Turtle()
celebration.speed(0)
celebration.color("yellow")
celebration.penup()
celebration.hideturtle()
celebration.goto(0, 100)

# Function to move paddle A up
def paddle_a_up():
    y = paddle_a.ycor()
    if y < 250:  # Prevent paddle from going off screen
        y += 20
        paddle_a.sety(y)

# Function to move paddle A down
def paddle_a_down():
    y = paddle_a.ycor()
    if y > -240:  # Prevent paddle from going off screen
        y -= 20
        paddle_a.sety(y)

# Function to move paddle B up
def paddle_b_up():
    y = paddle_b.ycor()
    if y < 250:
        y += 20
        paddle_b.sety(y)

# Function to move paddle B down
def paddle_b_down():
    y = paddle_b.ycor()
    if y > -240:
        y -= 20
        paddle_b.sety(y)

# Keyboard bindings
wn.listen()
wn.onkeypress(paddle_a_up, "w")  # Player A uses W and S keys
wn.onkeypress(paddle_a_down, "s")
wn.onkeypress(paddle_b_up, "Up")  # Player B uses up and down arrows
wn.onkeypress(paddle_b_down, "Down")

# Game loop
while True:
    wn.update()  # Updates the screen
    
    # Move the ball
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy)
    
    # Border checking
    if ball.ycor() > 290:
        ball.sety(290)
        ball.dy *= -1  # Reverse the y direction
        os.system("afplay bounce.wav&")  # Sound effect for Mac (remove if not working)
    
    if ball.ycor() < -290:
        ball.sety(-290)
        ball.dy *= -1
        os.system("afplay bounce.wav&")
    
    # Check if ball goes past paddles (score)
    if ball.xcor() > 390:
        celebration.write("ðŸŽ‰ Player A Scores! ðŸŽ‰", align="center", font=("Courier", 24, "normal"))
        wn.update()
        turtle.ontimer(lambda: celebration.clear(), 1000)  # Clear after 1 second
        ball.goto(0, 0)
        ball.dx *= -1
        score_a += 1
        pen.clear()
        pen.write("Player A: {}  Player B: {}".format(score_a, score_b), 
                 align="center", font=("Courier", 24, "normal"))
    
    if ball.xcor() < -390:
        celebration.write("ðŸŽ‰ Player B Scores! ðŸŽ‰", align="center", font=("Courier", 24, "normal"))
        wn.update()
        turtle.ontimer(lambda: celebration.clear(), 1000)
        ball.goto(0, 0)
        ball.dx *= -1
        score_b += 1
        pen.clear()
        pen.write("Player A: {}  Player B: {}".format(score_a, score_b), 
                 align="center", font=("Courier", 24, "normal"))
    
    # Paddle and ball collisions
    if (ball.xcor() > 340 and ball.xcor() < 350) and (ball.ycor() < paddle_b.ycor() + 50 and ball.ycor() > paddle_b.ycor() - 50):
        ball.setx(340)
        ball.dx *= -1
        os.system("afplay bounce.wav&")
    
    if (ball.xcor() < -340 and ball.xcor() > -350) and (ball.ycor() < paddle_a.ycor() + 50 and ball.ycor() > paddle_a.ycor() - 50):
        ball.setx(-340)
        ball.dx *= -1
        os.system("afplay bounce.wav&")
```
