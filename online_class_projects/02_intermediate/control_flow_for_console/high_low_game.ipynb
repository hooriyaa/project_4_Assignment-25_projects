```python
import random  # To generate random numbers

NUM_ROUNDS = 5  # Total rounds to play

def get_user_guess():
    """Function to get 'higher' or 'lower' input from the user."""
    while True:
        guess = input("Is your number higher or lower than the computer? (higher/lower): ").strip().lower()
        if guess in ["higher", "lower"]:
            return guess  # Valid input, return it
        print("Please enter only 'higher' or 'lower'!")  # Invalid input error

def play_round(round_num, score):
    """Function to play one round of the game."""
    user_number = random.randint(1, 100)
    computer_number = random.randint(1, 100)

    print(f"\nRound {round_num}")
    print(f"Your number: {user_number}")

    user_guess = get_user_guess()

    # Check if the user's guess was correct
    if (user_guess == "higher" and user_number > computer_number) or (user_guess == "lower" and user_number < computer_number):
        print(f"Correct! The computer's number was {computer_number}.")
        score += 1
    else:
        print(f"Oops! That was incorrect. The computer's number was {computer_number}.")

    print(f"Your score: {score}")
    return score  # Return updated score

def main():
    """Main function for the game."""
    print("üéÆ Welcome to the High-Low Game! üéÆ")
    print("-" * 30)

    score = 0  # Initial score is 0
    for round_num in range(1, NUM_ROUNDS + 1):
        score = play_round(round_num, score)

    # Final message based on the player's score
    print("\nThanks for playing!")
    print(f"Your final score: {score}")

    if score == NUM_ROUNDS:
        print("üëè Wow! You played perfectly! üéâ")
    elif score >= NUM_ROUNDS // 2:
        print("üëç Great job, you did well!")
    else:
        print("üò¢ Better luck next time!")

if __name__ == "__main__":
    main()
```
