```python
import random
import string

# Hangman visual representation with icons for different lives
lives_visual_dict = {
    7: "\n  +---+\n      |\n      |\n      |\n     ===  🟢🟢🟢🟢🟢🟢🟢",
    6: "\n  +---+\n  O   |\n      |\n      |\n     ===  🟢🟢🟢🟢🟢🟢🔴",
    5: "\n  +---+\n  O   |\n  |   |\n      |\n     ===  🟢🟢🟢🟢🟢🔴🔴",
    4: "\n  +---+\n  O   |\n /|   |\n      |\n     ===  🟢🟢🟢🟢🔴🔴🔴",
    3: "\n  +---+\n  O   |\n /|\  |\n      |\n     ===  🟢🟢🟢🔴🔴🔴🔴",
    2: "\n  +---+\n  O   |\n /|\  |\n /    |\n     ===  🟢🟢🔴🔴🔴🔴🔴",
    1: "\n  +---+\n  O   |\n /|\  |\n / \  |\n     ===  🟢🔴🔴🔴🔴🔴🔴",
    0: "\n  +---+\n  X   |\n /|\  |\n / \  |\n     ===  🔴🔴🔴🔴🔴🔴🔴",
}

# List of words to choose from
words = ["python", "developer", "frontend", "backend", "nextjs", "sanity", "machine", "learning"]

def get_valid_word(words):
    """Selects a random valid word from the list."""
    word = random.choice(words)  # Randomly selects a word
    while '-' in word or ' ' in word:  # Ensures no spaces or dashes in the word
        word = random.choice(words)
    return word.upper()  # Converts the word to uppercase

def hangman():
    """Main function to run the Hangman game."""
    word = get_valid_word(words)  # Selects a word for the game
    word_letters = set(word)  # Stores unique letters of the word
    alphabet = set(string.ascii_uppercase)  # Set of all uppercase letters
    used_letters = set()  # Tracks guessed letters
    lives = 7  # Maximum number of incorrect guesses allowed

    # Game loop: continues until all letters are guessed or lives are over
    while len(word_letters) > 0 and lives > 0:
        print(f"\nYou have {lives} lives left. Used letters: {' '.join(used_letters)}")
        print(lives_visual_dict[lives])  # Displays the current hangman stage with icons

        # Shows the word with guessed letters revealed
        word_display = [letter if letter in used_letters else '-' for letter in word]
        print("Current word:", ' '.join(word_display))

        # Takes user input
        user_letter = input("Guess a letter: 🎯 ").upper()

        # Checks if the guessed letter is valid
        if user_letter in alphabet - used_letters:
            used_letters.add(user_letter)
            if user_letter in word_letters:
                word_letters.remove(user_letter)  # Removes guessed letter from remaining letters
                print("✅ Correct guess!")
            else:
                lives -= 1  # Reduces a life for an incorrect guess
                print(f"❌ Your letter '{user_letter}' is not in the word.")
        elif user_letter in used_letters:
            print("⚠️ You have already used that letter. Try again.")
        else:
            print("⚠️ Invalid input. Please enter a valid letter.")

    # Game over scenarios
    if lives == 0:
        print(lives_visual_dict[lives])  # Displays final hangman stage
        print(f"\n💀 You lost! The word was '{word}'.")
    else:
        print(f"\n🎉 Congratulations! You guessed the word '{word}'! 🏆")

# Runs the game if the script is executed directly
if __name__ == "__main__":
    hangman()
```
